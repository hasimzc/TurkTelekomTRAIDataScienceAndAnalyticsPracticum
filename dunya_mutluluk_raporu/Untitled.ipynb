```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_path = 'World Happiness Report'
dataframe1=pd.read_csv(dataset_path+'/2015.csv')
dataframe2=pd.read_csv(dataset_path+'/2016.csv')
dataframe3=pd.read_csv(dataset_path+'/2017.csv')
dataframe4=pd.read_csv(dataset_path+'/2018.csv')
dataframe5=pd.read_csv(dataset_path+'/2019.csv')
concatenated_dataframes = pd.concat([dataframe1, dataframe2,dataframe3,dataframe4,dataframe5])
```

```python
# Question 1
# Which countries or regions rank highest in overall happiness and each of the six factors that contribute to happiness?



# Grouping by country and getting mean value
mean_values = concatenated_dataframes.groupby('Country').mean()

# Find the top 10 countries with highest overall happiness scores
top_10_countries = mean_values.sort_values("Happiness.Score", ascending=False).head(10)
print("Top 10 countries with highest overall happiness scores:\n")
print(top_10_countries[["Happiness.Score"]])

# Find the top 10 countries with highest scores for each of the six factors

print("\nTop 10 countries with highest scores for each of the six factors:\n")

for factor in mean_values.columns[[3,5,6,7,8,9]]:
    top_10_countries = mean_values.sort_values(factor, ascending=False).head(10)
    print(f"{factor}:\n")
    print(top_10_countries[[factor]])
    print()
```

```python
# Question 2
# How did country rankings or scores change between 2015-2016 and 2016-2017 reports?

# Find the change in overall happiness scores between 2015-2016 and 2016-2017


#Create a new column in each dataset to indicate the year
dataframe1['Year'] = 2015
dataframe2['Year'] = 2016
dataframe3['Year'] = 2017

# in dataframe 3 Happiness Score column is named as Happines.Score we should change name for correct merge process
dataframe3 = dataframe3.rename(columns={'Happiness.Score': 'Happiness Score'})


# Merge the two datasets on the 'Country' column, which contains the country names
merged_data15_16 = pd.merge(dataframe1, dataframe2, on='Country')
merged_data16_17 = pd.merge(dataframe2, dataframe3, on='Country')

# Calculate the difference in happiness score between 2015-2016 and 2016-2017 for each country
merged_data15_16['Score Difference'] = merged_data15_16['Happiness Score_y'] - merged_data15_16['Happiness Score_x']
merged_data16_17['Score Difference'] = merged_data16_17['Happiness Score_y'] - merged_data16_17['Happiness Score_x']







```

```python
# Print the countries that experienced the most significant increase or decrease in happiness score
print('Happiness score differences between 2015 and 2016\n')
print(merged_data15_16.sort_values(by='Score Difference', ascending=False).head(10)[['Country','Score Difference']])
print()
print(merged_data15_16.sort_values(by='Score Difference', ascending=True).head(10)[['Country','Score Difference']])
```

```python
print('Happiness score differences between 2016 and 2017\n')
print(merged_data16_17.sort_values(by='Score Difference', ascending=False).head(10)[['Country','Score Difference']])
print()
print(merged_data16_17.sort_values(by='Score Difference', ascending=True).head(10)[['Country','Score Difference']])
```

```python
# Question 3
# Has any country experienced a significant increase or decrease in happiness?

# calculate the difference in happiness scores between consecutive years
concatenated_dataframes["happiness_change"] = concatenated_dataframes.groupby("Country")["Happiness Score"].shift(periods=-1) - concatenated_dataframes["Happiness Score"]

# create a new column that indicates whether the change in happiness score is significant
concatenated_dataframes["change_significant"] = concatenated_dataframes["happiness_change"].apply(lambda x: 1 if x > 0.5 or x < -0.5 else 0)

# group by country and calculate the proportion of years in which the change in happiness score is significant
significant_change = concatenated_dataframes.groupby("Country")["change_significant"].mean()

# display the countries that experienced a significant increase or decrease in happiness
print(significant_change[significant_change > 0].index)
```
